```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('c:/Users/kkhan/OneDrive/Documents/GitHub/Celebal_Internship_07_14892/dataset/data.csv')
```

```python
data
```

```python
data.isna().sum()
```

```python
# Dropping the 'Unnamed: 32' column and 'id' column as they are not useful for analysis

data.drop(columns=['Unnamed: 32'],axis=1,inplace=True)
data.drop(columns=['id'],axis=1,inplace=True)
```

```python
# Convert categorical variable 'diagnosis' to numerical
data = data.replace({'diagnosis': {'M': 1, 'B': 0}})
```

```python
# Checking the correlation between features

plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True,annot_kws={"size": 8}, fmt='.000001f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Features')
plt.show()
```

```python
Y = data['diagnosis'] #This is the target variable.

X = data.drop(['diagnosis'], axis=1) #This is the feature set.
```

```python
#Checking the distribution of the target variable
print("Distribution of target variable:")
print(Y.value_counts(normalize=True))
```

```python
plt.figure(figsize=(10, 6)) 
sns.countplot(x='diagnosis', data=data, palette='viridis')
plt.title('Distribution of Diagnosis')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.show()
```

```python
# Checking the distribution of features

# Setting the style of seaborn
sns.set(style='whitegrid')

# Using histograms to visualize the distribution of each feature
for column in X.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(X[column], kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Checking the boxplot for each feature against the target variable
# Showing the outliers and distribution of each feature by diagnosis

sns.set(style='whitegrid')
for column in X.columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=Y, y=X[column], palette='Set2')
    plt.title(f'Boxplot of {column} by Diagnosis')
    plt.xlabel('Diagnosis')
    plt.ylabel(column)
    plt.show()
```

```python
# Checking the scatter plot for each feature against the target variable
# Showing the relationship between each feature and diagnosis

sns.set(style='whitegrid')
for column in X.columns:
    plt.figure(figsize=(10, 5))
    sns.scatterplot(x=X[column], y=Y, alpha=0.6)
    plt.title(f'Scatter Plot of {column} vs Diagnosis')
    plt.xlabel(column)
    plt.ylabel('Diagnosis')
    plt.show()
```
