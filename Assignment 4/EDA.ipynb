*Separate Movies and TV Shows data*

*Convert 'duration' to numeric values for Movies and TV Shows*

```python
data_genres = data[['listed_in', 'release_year']].dropna() #data for genres and release year
```

**Plotting the trends of top genres over time**



```python
data_genres = data_genres.assign(genre = data_genres['listed_in'].str.split(', ')).explode('genre') # Explode the 'genre' column to have one genre per row

genre_trends = data_genres.groupby(['genre', 'release_year']).size().reset_index(name='count') #genre trends over years
genre_trends = genre_trends[genre_trends['release_year'] > 2000] # Filter for years after 2000

top_genres = genre_trends.groupby('genre')['count'].sum().sort_values(ascending=False).head(10).index # Get top 10 genres by total count
filtered_trends = genre_trends[genre_trends['genre'].isin(top_genres)] # Filter trends for top genres

plt.figure(figsize=(14, 8))
sns.lineplot(data=filtered_trends, x='release_year', y='count', hue='genre', marker='o')
plt.title("Top genres over time on Netfix(2000 onwards)")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.legend(title='Genre')
plt.grid(True)
plt.tight_layout()
plt.show()
```

**Data for Country And Genre**

```python
data_country_genre = data[['country', 'listed_in']].dropna() 
```

*Split the 'country' and 'listed_in' columns into lists*

*This allows us to analyze the relationship between countries and genres and explode the lists to create a row for each country-genre pair*

```python
data_country_genre['country'] = data_country_genre['country'].str.split(', ')
data_country_genre['genre'] = data_country_genre['listed_in'].str.split(', ')
data_country_genre = data_country_genre.explode('country')
data_country_genre = data_country_genre.explode('genre')

# Group by country and genre to count occurrences
country_genre_counts = data_country_genre.groupby(['country', 'genre']).size().reset_index(name='count')

# Create a pivot table to summarize counts of genres by country
pivot_table = country_genre_counts.pivot(index='country', columns='genre', values='count').fillna(0)

# Normalize the pivot table to get proportions
top_countries = data['country'].dropna().str.split(', ').explode().value_counts().head(10).index
pivot_table = pivot_table.loc[pivot_table.index.isin(top_countries)]
```

**Plotting The Heatmap**

```python
plt.figure(figsize=(14, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', linewidths=0.5 ,annot=False)
plt.title("Heatmap of Country vs Genre")
plt.xlabel("Genre")
plt.ylabel("Country")
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
